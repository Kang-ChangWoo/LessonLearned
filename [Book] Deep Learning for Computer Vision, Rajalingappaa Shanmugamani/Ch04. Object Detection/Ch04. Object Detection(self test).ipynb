# 01. Sin, Cos 0에서 10까지 그리기, 레전드 및 각 라벨을 그리고 타이틀을 설정하기 cos는 점선으로 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
# y3 = np.tan(x)

plt.plot(x, y1, label ="Sign graph")
plt.plot(x, y2, linestyle = '--', label = "Cosign graph")
# plt.plot(x, y3, linestyle = '-.', label = "Tangent graph")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sin and Cos graph")

plt.legend()
plt.show()
```
